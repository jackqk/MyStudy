# 一、创建Series

## 从list创建
自动帮你生成从0开始的整数索引

```
#注意dtype是object
animals = ['tiger', 'bear', 'moose']
pd.Series(animals)
```

```
numbers = [1,2,3]
pd.Series(numbers)
```

```
s = pd.Series(['tiger', 'bear', 'moose'], index = ['India', 'America', 'Cannad'])
s
```

## 从dictionary创建
可以给索引命名

```
sports = {'Archery' : 'Bhutan',
         'Golf' : 'Scotland',
         'Sumo' : 'Japan',
         'Taekwondo' : 'South Koear' }
s = pd.Series(sports)
s
```

```
s.index
```

## pandas对python的None处理

```
#自动转为Object
animals = ['tiger', 'bear', None]
pd.Series(animals)
```

```
#自动转为np.nan
numbers = [1,2,None]
pd.Series(numbers)
```

```
#这个NaN十分有趣
import numpy as np

print(np.nan == None)
print(np.nan == np.nan)
  print(np.isnan(np.nan))
```

# 二、查询/选择

## 像list，dict那样查询

```
sports = {'Archery' : 'Bhutan',
         'Golf' : 'Scotland',
         'Sumo' : 'Japan',
         'Taekwondo' : 'South Koear' }
s = pd.Series(sports)
```

```
#位置查询（如同list）
#这之间有一个意外，就是当index为整数时，那么这种查询就会失效
s[0]
```

```
#index name
s['Golf']
```

## iloc、loc查询
loc如果传入的没有的列，这会创建一个新的列

```
#查第几行
s.iloc[1]
```

```
#列名查找
s.loc['Golf']
```

## 当索引是整数是时出现的问题

```
#当索引是整数时，
sports = {99: 'Bhutan',
          100: 'Scotland',
          101: 'Japan',
          102: 'South Korea'}
s = pd.Series(sports)
```

```
#出错（不能以list方式来查找）
s[0]
```

```
s.iloc[0]
```

## 选择多个index

```
original_sports = pd.Series({'Archery': 'Bhutan',
                             'Golf': 'Scotland',
                             'Sumo': 'Japan',
                             'Taekwondo': 'South Korea'})
cricket_loving_countries = pd.Series(['Australia',
                                      'Barbados',
                                      'Pakistan',
                                      'England'], 
                                   index=['Cricket',
                                          'Cricket',
                                          'Cricket',
                                          'Cricket'])
all_countries = original_sports.append(cricket_loving_countries)
```

```
# 拥有相同index
all_countries['Cricket']
```

```
# 切片,可以选择多个Index
all_countries[0:3]
```

# 三、Operation

## 遍历
在遍历前，想想是否能够使用向量化，因为更快

**方法一：**<br>

```
#默认遍历value
#可以用s.keys()方法，来指定遍历什么
s = pd.Series([100.0, 110.0, 130.0,22.0])
for item in s:
    print(item)
```

**方法二：**

```
#key value都遍历出来
for key, value in s.iteritems():
  s.at[key] = value + 20
s
```

## 数学运算

```
s = pd.Series([100.0, 110.0, 130.0,22.0])
s
```

```
np.sum(s)
```

```
s += 2
s
```

```
s = s ** 2
s
```

## 关于速度

```
s = pd.Series(np.random.randint(0, 1000, 10000))
s.head()
```

```
%%timeit -n 100
#下面的执行100次，取最佳时间
summary = 0
for item in s:
  summary += item
```

```
%%timeit -n 100
summary = np.sum(s)
```
